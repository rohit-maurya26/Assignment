```python
#Q2
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal over Time for Different Events and Regions")
plt.show()
```

```python
#Q3
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Age Distribution by Passenger Class")

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Fare Distribution by Passenger Class")

plt.tight_layout()
plt.show()
```

```python
#Q4
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.title("Price Distribution of Diamonds by Cut")
plt.show()
```

```python
#Q5
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
sns.pairplot(iris, hue='species')
plt.suptitle("Pair Plot of Iris Dataset by Species", y=1.02)
plt.show()
```

```python
#Q6
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, annot=True, fmt=".0f", cmap="YlGnBu") # Changed fmt to ".0f"
plt.title("Number of Passengers per Month and Year")
plt.ylabel("Month")
plt.xlabel("Year")
plt.show()
```
