```python
#Q2
#When an exception is not handled, the program terminates and displays a traceback message, showing the line where the exception occurred and the type of exception.
#Example:
import logging

logging.basicConfig(filename="error.log" ,level=logging.ERROR)

try:
    result = 10 / 0
except ZeroDivisionError as e:
    logging.error("An exception occurred: %s", e)
```

```python
#Q3
#The try and except statements are used to catch and handle exceptions in Python.
#Example:
import logging

logging.basicConfig(filename="error.log" , level=logging.ERROR)

try:
    a = [1, 2, 3]
    print(a[3])
except IndexError as e:
    logging.error("An exception occurred: %s", e)

```

```python
#Q4
#try and else: 
The else block executes if no exceptions are raised in the try block.
#EX:
import logging

logging.basicConfig(filename="error.log" ,level=logging.INFO)

try:
    num = 10 / 2
except ZeroDivisionError as e:
    logging.error("An exception occurred: %s", e)
else:
    logging.info("No exceptions occurred. The result is %d", num)


    
#finally: 
    The finally block executes regardless of whether an exception is raised or not.
#Ex:
import logging

logging.basicConfig(filename="error.log" ,level=logging.INFO)

try:
    num = 10 / 0
except ZeroDivisionError as e:
    logging.error("An exception occurred: %s", e)
finally:
    logging.info("This block is executed no matter what.")


#raise:
The raise statement is used to manually raise an exception.

#Ex:
import logging

logging.basicConfig(filename="error.log" , level=logging.ERROR)

def check_value(x):
    if x < 0:
        raise ValueError("Negative values are not allowed")
    return x

try:
    value = check_value(-1)
except ValueError as e:
    logging.error("An exception occurred: %s", e)

```

```python
#Q5
Custom exceptions are user-defined exceptions that extend Python's built-in Exception class. They allow for more specific error handling and can provide more descriptive error messages.

#Why we need Custom Exceptions:

To give more specific error messages.

To handle specific error situations more effectively.

To make the code easier to understand and maintain.

#ex:
    import logging

logging.basicConfig(filename="error.log" , level=logging.ERROR)

class NegativeNumberError(Exception):
    pass

def check_positive(number):
    if number < 0:
        raise NegativeNumberError("Negative numbers are not allowed")
    return number

try:
    check_positive(-5)
except NegativeNumberError as e:
    logging.error("An error occurred: %s", e)

```
