```python
#Q1
Matplotlib is a Python library used to create graphs and charts. 
It helps visualize data in a clear way. 
Five types of plots you can make with Pyplot are:
- Line Plot
- Bar Plot
- Scatter Plot
- Histogram
- Box Plot
```

```python
#Q2
# A scatter plot shows individual points on a graph. 
# It helps see the relationship between two sets of values.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of X and Y")
plt.show()
```

```python
#Q3
# The subplot() function in Matplotlib is used to create multiple plots within a single figure. 
# It allows you to arrange several axes in a grid layout.

import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 40, 20, 60, 70, 10])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Plot 1")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Plot 2")

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Plot 3")

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Plot 4")

plt.tight_layout()
plt.show()
```
