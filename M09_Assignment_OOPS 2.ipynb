```python
#Q1
#Class
A class is like a blueprint or template.For example if we want to build several houses with the same design. The blueprint is the class. It defines the structure and features of the houses, but it doesn't create any actual houses by itself.
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
    
    def display_info(self):
        print(f"{self.year} {self.brand} {self.model}")
#Object
An object is an actual instance of the class. Using our house analogy, an object is one of the houses built using the blueprint. Each house (object) has the same design (class), but they are individual houses.
#Ex
car1 = Car("Toyota", "Corolla", 2020)
car2 = Car("Honda", "Civic", 2019)

car1.display_info()  
car2.display_info()  


```

```python
#Q2
Encapsulation::     Bundles the data and Hides the internal details and only exposes what's necessary. Like a TV remote, we use it without knowing its inner workings.

Inheritance::    Allows one class to inherit methods and variables from another. Like a child inheriting traits from their parents.

Polymorphism::      Lets methods do different things based on the object calling them. Like a word having different meanings in different contexts.

Abstraction::     Simplifies complex systems by focusing only on essential features. Like using a car without needing to understand the engine.
```

```python
#Q3
The __init__() function in Python is called a constructor. It is used to initialize the attributes of an object when it is created. Think of it as a setup function that prepares the object with initial values.

class Car:
   def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

car1 = Car("Toyota", "Corolla", 2020)
print(car1.brand) 
```

```python
#Q4
The self keyword in OOPS in Python is used to refer to the instance of the class itself.
#EX
class Cat:
    def __init__(self, name):
        self.name = name

    def sound(self):
        print(f"{self.name} says meow ")

dog1 = Cat("Tom")
dog1.sound() 
```

```python
#Q5
Inheritance allows a new class to take on the properties and methods of an existing class. It's like inheriting features from a parent.
#Types of Inheritence
#1-Single Inheritance:
A class inherits from one parent class.
#Ex
class Animal:
    def eat(self):
        print("Eating...")

class Dog(Animal):
    pass

dog = Dog()
dog.eat()


#2-Multiple Inheritance:
A class inherits from more than one parent class.
#Ex
class Father:
    def height(self):
        print("Tall")

class Mother:
    def intelligence(self):
        print("Smart")

class Child(Father, Mother):
    pass

child = Child()
child.height()         
child.intelligence()  

#3-Multilevel Inheritance:
A class is derived from a class that is also derived from another class.
#Ex
class Grandparent:
    def legacy(self):
        print("Legacy")

class Parent(Grandparent):
    pass

class Child(Parent):
    pass

child = Child()
child.legacy()  

#4-Hierarchical Inheritance:
Multiple classes inherit from a single parent class.
#Ex
class Vehicle:
    def start(self):
        print("Starting...")

class Car(Vehicle):
    pass

class Bike(Vehicle):
    pass

car = Car()
bike = Bike()
car.start()  
bike.start() 

#


```
